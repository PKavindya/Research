# Parkinson's vs Parkinsonian 

```python
import pandas as pd

df = pd.read_csv('parkinsons and parkinsonian.csv',index_col=False)
df.head()
```

```python
df.rename(columns = {'ID':'Patient ID'}, inplace = True)
```

```python
X = df.drop(['clinical_diagnosis'], axis=1)
```

```python
X
```

```python
Y = df['clinical_diagnosis']
```

```python
Y
```

```python
#normalized data with Standard Scaler

from sklearn import preprocessing
scaler = preprocessing.StandardScaler()

scaler.fit(X)
Xs = scaler.transform(X)
Xs
```

```python
# convert array data set got for X into a dataframe

Xdf = pd.DataFrame(Xs, columns = X.columns)
Xdf
```

```python
#Selecting 100 features using mutual feature selection

import matplotlib.pyplot as plt

from sklearn.feature_selection import mutual_info_classif

from sklearn.feature_selection import SelectKBest

# apply score function and use best 100 features from dataset (k=100)

df = SelectKBest(score_func=mutual_info_classif, k=100)

df.fit(Xdf,Y)

mi_score = pd.Series(df.scores_, index=X.columns)

mi_score.sort_values(ascending=False).plot.bar(figsize=(6, 4))

plt.show()

print("\nFirst 100 rows of feature set after feature selection")

X_selected = df.fit_transform(Xdf, Y)

X_selected = pd.DataFrame(X_selected,columns=df.get_feature_names_out())
```

```python
X_selected.to_csv("Selected_100_features_from_parkinson's_vs_parkinsonian_data_set.csv", encoding="utf-8", index = False)
```

```python
Y.to_csv("y(parkinson's and parkinsonian).csv", encoding="utf-8", index = False)
```

```python
#balancing data using SMOTE

from collections import Counter
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_selected, Y)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
acc = cross_val_score(rf, X_res, y_res, cv=10)
print("Accuracy RF: %0.2f (+/- %0.2f)" % (acc.mean(), acc.std() * 2))

# Support Vector Machine
from sklearn.svm import SVC
svm = SVC()
acc = cross_val_score(svm, X_res, y_res, cv=10)
print("Accuracy SVM: %0.2f (+/- %0.2f)" % (acc.mean(), acc.std() * 2))

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
acc = cross_val_score(knn, X_res, y_res, cv=10)
print("Accuracy KNN: %0.2f (+/- %0.2f)" % (acc.mean(), acc.std() * 2))
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_res, y_res, test_size=0.25)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

# Support Vector Machine
from sklearn.svm import SVC
svm = SVC()

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
param_grid_rf = {
    'n_estimators': [100, 200, 300]
}
param_grid_svc = {
    'C': [0.1,1, 10, 100],
    'gamma': [1,0.1,0.01,0.001],
    'kernel': ['rbf', 'poly', 'sigmoid']
}
param_grid_knn = {
    'n_neighbors':[i for i in range(1,20)]
}
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV

# Random Forest
gridSearchRF = GridSearchCV(estimator=rf, param_grid=param_grid_rf, scoring='accuracy', n_jobs=-1, cv=5)
gridSearchRF.fit(X_train, y_train.ravel())

y_pred = gridSearchRF.predict(X_test)
target_names = ['class 0', 'class 1']

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
# K-Nearest Neighbors

gridSearchKNN = GridSearchCV(estimator=knn, param_grid=param_grid_knn, scoring='accuracy', n_jobs=-1, cv=5)
gridSearchKNN.fit(X_train, y_train.ravel())

y_pred = gridSearchKNN.predict(X_test)
target_names = ['class 0', 'class 1']

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
# Support Vector Machine

gridSearchSVM = GridSearchCV(estimator=svm, param_grid=param_grid_svc, scoring='accuracy', n_jobs=-1, cv=5)
gridSearchSVM.fit(X_train, y_train.ravel())

y_pred = gridSearchSVM.predict(X_test)
target_names = ['class 0', 'class 1']

print(classification_report(y_test, y_pred, target_names=target_names))
```
